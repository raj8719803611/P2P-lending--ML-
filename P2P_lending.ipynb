# Ridge and Lasso

Ridge and Lassoare some of the simple techniques to reduce model complexity and prevent over-fitting which may 
result from simple linear regression.
ridge shrinks the coefficients and it helps to reduce the model complexity and multi-collinearity. 

```python
from sklearn import linear_model

regr_lasso = linear_model.Lasso(alpha=0.1)

regr_ridge = linear_model.Ridge(alpha=1.0)
```

```python
ridge=regr_ridge.fit(xtrainscale,ytrain)

rpred=ridge.predict(xtestscale)

rmseR=np.sqrt(mean_squared_error(rpred,ytest))

print(rmseR)

print(r2_score(rpred,ytest))
```

```python
lasso=regr_lasso.fit(xtrainscale,ytrain)

lpred=lasso.predict(xtestscale)

rmsel=np.sqrt(mean_squared_error(ytest,lpred))

print(rmsel)

print(r2_score(lpred,ytest))
```

### Random Forest Regressor

The fundamental idea behind a random forest is to combine the predictions made by many decision trees into a single model. 
Individually, predictions made by decision trees may not be accurate but combined together, the predictions will be closer 
to the true value on average.

```python

from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 15 decision trees
rf = RandomForestRegressor(n_estimators = 15, random_state = 0)
# Train the model on training data
rf.fit(xtrainscale,ytrain)
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(xtestscale)
# Calculate the absolute errors
errors = abs(predictions - ytest)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / ytest)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
rmse=np.sqrt(mean_squared_error(predictions,ytest))
print(rmse)
print(r2_score(predictions,ytest))
```

Conclusion on Best Regression model:
We would be considering RMSE as performance measure for our regression problem. It gives us an idea of how much error the model 
makes in its predictions. All the models performed well on this dataset, but to conclude the best models are lasso 
and Knn regressor with RMSE as 1.

```python
from sklearn.model_selection import KFold
from sklearn.metrics import roc_curve, auc
kf=KFold(n_splits=5,shuffle=True,random_state=0)
for model, name in zip([logit,classifier,rf_model,knn_classifier],['LR','NB','RF','KNN']):
    roc_auc=[]
    for train,test in kf.split(x1,y1):
        Xtrain,Xtest = x1[train,:].values,x1[test,:].values
        Ytrain,Ytest = y1[train].values,y1[test].values
        model.fit(Xtrain,Ytrain)
        Y_predict=model.predict(Xtest)
        #cm=metrics.confusion_matrix(Ytest,Y_predict)
        fpr,tpr,_ = roc_curve(Ytest,Y_predict)
        roc_auc.append(auc(fpr,tpr))
    print("AUC score: %0.02f (+/-)%0.5f)[%s]"% (np.mean(roc_auc),
                                               np.var(roc_auc,ddof=1),name))
```
